# ¿Qué son los Entity Embeddings y cómo se relacionan con las variables categóricas?

Entity Embeddings es una técnica para aplicar deep learning a datos tabulares. Implica representar los datos categóricos de una entidad de sistemas de información con múltiples dimensiones.

Las Embeddings se crean principalmente como una forma de representar algo en formato vectorial, para ser utilizado en un modelo de aprendizaje profundo. Es útil tenerlos, ya que nos permite capturar y enviar solo la información más destacada necesaria de los datos originales a nuestro modelo.

se relacionan con las variables categóricas ya que dependiendo de su configuración representan variables categóricas de forma compacta y continua.
